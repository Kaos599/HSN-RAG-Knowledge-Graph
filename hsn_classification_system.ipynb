# 2. Data Processing: HSN Hierarchy

```python
def extract_hsn_data(pdf_path):
    data = []

    current_chapter_code = None
    current_chapter_desc = None
    current_heading_code = None
    current_heading_desc = None
    current_subheading_code = None
    current_subheading_desc = None

    logger.info(f"Starting PDF extraction from {pdf_path}...")

    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            tables = page.extract_tables()
            for table in tables:
                for row in table:
                    # Clean row values
                    row = [str(cell).strip() if cell else "" for cell in row]

                    # Skip headers and empty rows
                    if 'ITC(HS)\nCode' in row or 'ITC(HS)' in row[1]: continue
                    if not any(row): continue
                    if len(row) < 3: continue
                    # Code column must be digit-like (ignoring newlines)
                    clean_code = row[1].replace('\n', '').strip()
                    if not clean_code.isdigit(): continue

                    # Extract basic fields
                    code = clean_code
                    description = row[2].replace('\n', ' ').strip()
                    policy = row[3].replace('\n', ' ').strip() if len(row) > 3 else ""

                    # Hierarchy Logic
                    if len(code) == 2:
                        current_chapter_code = code
                        current_chapter_desc = description
                        current_heading_code = None; current_heading_desc = None
                        current_subheading_code = None; current_subheading_desc = None

                    elif len(code) == 4:
                        current_heading_code = code
                        current_heading_desc = description
                        current_subheading_code = None; current_subheading_desc = None

                    elif len(code) == 6:
                        current_subheading_code = code
                        current_subheading_desc = description

                    elif len(code) == 8:
                        # Handle implicit subheading updates (merged cells in PDF)
                        sub_code = code[:6]
                        item_desc = description

                        if sub_code != current_subheading_code:
                            # Subheading changed without a 6-digit row
                            # Attempt to split "Subheading : Item" or "Subheading -- Item"
                            split_match = re.search(r'(\s*:\s*-{0,2}\s*|\s+--\s+)', description)

                            if split_match:
                                sep_start, sep_end = split_match.span()
                                new_sub_desc = description[:sep_start].strip()
                                item_desc = description[sep_end:].strip()
                                current_subheading_code = sub_code
                                current_subheading_desc = new_sub_desc
                            else:
                                # Fallback
                                current_subheading_code = sub_code
                                current_subheading_desc = description

                        entry = {
                            "ChapterNumber": current_chapter_code,
                            "HSN Code": code,
                            "Description": item_desc,
                            "FinalHSN": policy,
                            "Chapter_Description": current_chapter_desc,
                            "Heading_Description": current_heading_desc,
                            "Subheading_Description": current_subheading_desc
                        }
                        data.append(entry)

    df = pd.DataFrame(data)
    logger.info(f"Extraction complete. Found {len(df)} records.")
    return df

# Run Extraction
pdf_file = "Trade_Notice_First_50_Pages.pdf"
if os.path.exists(pdf_file):
    df = extract_hsn_data(pdf_file)
    print(df.head())
else:
    raise FileNotFoundError(f"{pdf_file} not found. Please upload the dataset.")
```

# 2.2 Document Structure Creation

```python
#documents for the RAG system with enhanced content
def create_documents(df):
    documents = []
    for _, row in df.iterrows():
        # Handle None values safely
        chapter_desc = str(row['Chapter_Description']) if row['Chapter_Description'] else ""
        heading_desc = str(row['Heading_Description']) if row['Heading_Description'] else ""
        subheading_desc = str(row['Subheading_Description']) if row['Subheading_Description'] else ""
        item_desc = str(row['Description']) if row['Description'] else ""

        # Determine category markers based on heading code for semantic distinction
        heading_code = row['HSN Code'][:4]
        category_marker = ""
        semantic_boost = ""

        # Add strong category markers to create semantic distance
        if heading_code == "4001":
            category_marker = "[NATURAL RUBBER - NOT SYNTHETIC]"
            semantic_boost = "natural rubber, natural gum, natural latex, from rubber trees"
        elif heading_code == "4002":
            category_marker = "[SYNTHETIC RUBBER - NOT NATURAL]"
            semantic_boost = "synthetic rubber, man-made rubber, artificial rubber, chemical rubber"
        elif heading_code == "4005":
            category_marker = "[COMPOUNDED RUBBER - MIXED/PROCESSED]"
            semantic_boost = "compounded rubber, mixed rubber, processed rubber"
        elif heading_code == "4017":
            category_marker = "[HARDENED RUBBER - EBONITE/VULCANITE]"
            semantic_boost = "hardened rubber, ebonite, vulcanite, hard rubber products"

        # Add specific description markers
        desc_markers = ""
        if "prevulcanis" in item_desc.lower() or "pre-vulcanis" in item_desc.lower():
            desc_markers = "[PREVULCANISED - PRE-TREATED LATEX]"
        elif "latex" in subheading_desc.lower() or "latex" in item_desc.lower():
            desc_markers = "[LATEX FORM - LIQUID RUBBER]"

        # Enhanced content with emphasis on key terms for better retrieval
        content = f"""
{category_marker} {desc_markers}
HSN Code: {row['HSN Code']}
Chapter ({row['ChapterNumber']}): {chapter_desc}
Heading ({row['HSN Code'][:4]}): {heading_desc}
Subheading ({row['HSN Code'][:6]}): {subheading_desc}
Specific Description: {item_desc}
Export Policy: {row['FinalHSN']}

Category: {semantic_boost}
Keywords: {chapter_desc.lower()}, {heading_desc.lower()}, {subheading_desc.lower()}, {item_desc.lower()}
""".strip()

        doc = Document(
            page_content=content,
            metadata={
                "hsn_code": row['HSN Code'],
                "chapter": row['ChapterNumber'],
                "heading": row['HSN Code'][:4],
                "subheading": row['HSN Code'][:6],
                "description": item_desc,
                "full_context": f"{chapter_desc} > {heading_desc} > {subheading_desc}",
                "category": category_marker
            }
        )
        documents.append(doc)
    return documents

documents = create_documents(df)
print(f"Created {len(documents)} documents for RAG.")
```

# 3. Knowledge Graph Construction
This section builds a directed graph representing the HSN hierarchy (Chapter -> Heading -> Subheading -> Code) using NetworkX.


```python

# Using NetworkX to build the hierarchy and PyVis to visualize it.
def build_hsn_graph(df):
    G = nx.DiGraph()

    # 1. Add Chapters (Level 0)
    chapters = df[['ChapterNumber', 'Chapter_Description']].drop_duplicates().dropna()
    for _, row in chapters.iterrows():
        node_id = row['ChapterNumber']
        label = f"Ch {node_id}: {row['Chapter_Description'][:20]}..."
        G.add_node(node_id, label=label, level=0, color='#FF6B6B', title=row['Chapter_Description'])

    # 2. Add Headings (Level 1)
    df['Heading_Code'] = df['HSN Code'].str[:4]
    headings = df[['Heading_Code', 'Heading_Description', 'ChapterNumber']].drop_duplicates().dropna()
    for _, row in headings.iterrows():
        node_id = row['Heading_Code']
        label = f"{node_id}: {row['Heading_Description'][:20]}..."
        G.add_node(node_id, label=label, level=1, color='#4ECDC4', title=row['Heading_Description'])
        G.add_edge(row['ChapterNumber'], node_id)

    # 3. Add Subheadings (Level 2)
    df['Subheading_Code'] = df['HSN Code'].str[:6]
    subheadings = df[['Subheading_Code', 'Subheading_Description', 'Heading_Code']].drop_duplicates().dropna()
    for _, row in subheadings.iterrows():
        node_id = row['Subheading_Code']
        label = f"{node_id}: {str(row['Subheading_Description'])[:20]}..."
        G.add_node(node_id, label=label, level=2, color='#45B7D1', title=str(row['Subheading_Description']))
        G.add_edge(row['Heading_Code'], node_id)

    # 4. Add Specific Codes (Level 3)
    for _, row in df.iterrows():
        node_id = row['HSN Code']
        label = f"{node_id}: {row['Description'][:20]}..."
        G.add_node(node_id, label=label, level=3, color='#96CEB4', title=f"{row['Description']} (Policy: {row['FinalHSN']})")
        G.add_edge(row['Subheading_Code'], node_id)

    return G

G = build_hsn_graph(df)
print(f"✓ Graph built with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")
```

If the interactive graph above does not render (e.g., due to browser security settings or missing data), this static plot confirms the graph structure.


```python

# Static visualization using Matplotlib (Fallback)
import matplotlib.pyplot as plt

def visualize_static_graph(G):
    if G.number_of_nodes() == 0:
        print("⚠️ Graph is empty.")
        return

    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, seed=42)  # Force-directed layout

    # Draw nodes
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color='lightblue', alpha=0.7)

    # Draw edges
    nx.draw_networkx_edges(G, pos, width=0.5, alpha=0.5, arrows=True)

    # Draw labels (only for higher levels to avoid clutter)
    # Filter labels for Chapters and Headings only
    labels = {n: G.nodes[n]['label'] for n in G.nodes if G.nodes[n].get('level', 3) < 2}
    nx.draw_networkx_labels(G, pos, labels, font_size=8)

    plt.title("HSN Knowledge Graph (Static View)")
    plt.axis('off')
    plt.show()

print("Generating static visualization...")
visualize_static_graph(G)
```

# 4.2 Intelligent Query Processing - Part 1: Disambiguation Function

```python
# Retrieves results and determines if user query is ambiguous with improved filtering

def retrieve_with_disambiguation(query, vector_store, gemini_client, threshold=0.12, k=15):
    """
    Retrieve results and check for ambiguity with hierarchical filtering.
    Enhanced with keyword-based pre-filtering and re-ranking.
    Returns a dict with type: 'direct_answer', 'disambiguation', or 'no_result'
    """
    # 1. Retrieve more results initially for better filtering
    results = vector_store.similarity_search_with_score(query, k=k)

    if not results:
        return {"type": "no_result", "message": "No relevant HSN codes found."}

    # 1.5 KEYWORD-BASED PRE-FILTERING AND RE-RANKING
    query_lower = query.lower()

    # Determine preferred heading based on query keywords
    preferred_heading = None
    must_match_keywords = []

    # Check for natural vs synthetic rubber
    if "natural rubber" in query_lower or ("natural" in query_lower and "rubber" in query_lower):
        preferred_heading = "4001"
    elif "synthetic rubber" in query_lower or ("synthetic" in query_lower and "rubber" in query_lower):
        preferred_heading = "4002"
    elif "natural" in query_lower and "latex" in query_lower:
        preferred_heading = "4001"  # Natural latex is heading 4001

    # Check for specific keywords
    if "prevulcanis" in query_lower or "pre-vulcanis" in query_lower or "prevulcanised" in query_lower:
        must_match_keywords.append("prevulcanis")
    if "latex" in query_lower:
        must_match_keywords.append("latex")

    # Re-rank results: boost matching headings, penalize non-matching
    reranked_results = []
    for doc, score in results:
        heading = doc.metadata['heading']
        desc_lower = doc.metadata['description'].lower()
        subheading_code = doc.metadata['subheading']
        full_content_lower = doc.page_content.lower()

        # Apply boost/penalty
        adjusted_score = score

        # Boost if heading matches preferred
        if preferred_heading and heading == preferred_heading:
            adjusted_score -= 0.15  # Significant boost (lower score = better)
        elif preferred_heading and heading != preferred_heading:
            # Check for conflicting headings (natural vs synthetic)
            if preferred_heading == "4001" and heading in ["4002", "4005", "4017"]:
                adjusted_score += 0.30  # Strong penalty for wrong category
            elif preferred_heading == "4002" and heading == "4001":
                adjusted_score += 0.30  # Strong penalty

        # Boost if must-match keywords are in description or content
        for kw in must_match_keywords:
            if kw in desc_lower or kw in full_content_lower:
                adjusted_score -= 0.12  # Boost for keyword match
            else:
                adjusted_score += 0.08  # Penalty for missing keyword

        # Special case: "prevulcanised" should strongly prefer 40011010 (subheading 400110)
        if "prevulcanis" in query_lower and subheading_code == "400110":
            adjusted_score -= 0.25  # Strong boost for correct subheading

        # Special case: "natural rubber latex" should prefer subheading 400110
        if "natural" in query_lower and "latex" in query_lower and subheading_code == "400110":
            adjusted_score -= 0.20  # Strong boost

        reranked_results.append((doc, adjusted_score, score))  # Keep original score for reference

    # Sort by adjusted score
    reranked_results.sort(key=lambda x: x[1])
    results = [(doc, orig_score) for doc, adj_score, orig_score in reranked_results]

    # 2. Filter for relevance - group by subheading to find natural clusters
    subheading_groups = {}
    for doc, score in results:
        subheading = doc.metadata['subheading']
        if subheading not in subheading_groups:
            subheading_groups[subheading] = []
        subheading_groups[subheading].append((doc, score))

    # Find the subheading from the top result (already re-ranked)
    best_subheading = results[0][0].metadata['subheading']

    # Filter to only the best subheading's codes
    filtered_results = subheading_groups.get(best_subheading, [])[:5]

    if not filtered_results:
        filtered_results = results[:5]  # Fallback

    top_doc, top_score = filtered_results[0]

    # 3. Check for Ambiguity within the filtered results
    if len(filtered_results) >= 2:
        second_doc, second_score = filtered_results[1]
        score_diff = abs(top_score - second_score)

        # Trigger disambiguation if scores are close
        if score_diff < threshold:
            # Ambiguous!
            options = [top_doc, second_doc]

            # Generate clarifying question
            prompt_text = f"""You are an HSN code classification expert. A user asked: "{query}".
I found two likely HSN codes, but they are very similar.

Option 1:
{top_doc.page_content}

Option 2:
{second_doc.page_content}

Generate a SINGLE, short question to help the user distinguish between these two options.
Focus on the key difference (e.g. material, state, usage, pre-treatment)."""

            try:
                contents = [
                    types.Content(
                        role="user",
                        parts=[types.Part.from_text(text=prompt_text)],
                    ),
                ]

                # Use the helper for config
                generate_content_config = get_gen_config(GEMINI_MODEL, thinking_level="LOW")

                response = gemini_client.models.generate_content(
                    model=GEMINI_MODEL,
                    contents=contents,
                    config=generate_content_config,
                )

                question = response.text

            except Exception as e:
                logger.warning(f"Gemini API failed: {str(e)}")
                question = "Please specify which option better describes your item?"

            return {
                "type": "disambiguation",
                "question": question,
                "options": [
                    {"code": d.metadata['hsn_code'], "desc": d.metadata['description']}
                    for d in options
                ]
            }

    # 4. Direct Answer (High confidence or clear winner)
    return {
        "type": "direct_answer",
        "hsn_code": top_doc.metadata['hsn_code'],
        "document": top_doc,
        "score": top_score
    }
```

# 4.2 Intelligent Query Processing - Part 2: HSNClassifier Class

```python
# Main classifier class that handles user queries and returns HSN codes

class HSNClassifier:
    def __init__(self, vector_store, graph, gemini_client):
        self.vector_store = vector_store
        self.graph = graph
        self.gemini_client = gemini_client

    def query(self, user_query):
        """
        Main query method that processes user input and returns HSN code information.
        """
        # 0. Check for Direct HSN Code Lookup (TC5) - MOVED UP FOR PRIORITY
        hsn_match = re.search(r'\b\d{8}\b', user_query)
        if hsn_match:
            code = hsn_match.group(0)
            if code in self.graph:
                 # Fetch details from graph
                 node_data = self.graph.nodes[code]
                 desc = node_data.get('title', 'No description')
                 path_str = self.get_hierarchy_text(code)

                 # Extract policy from title if stored there, or from metadata if available
                 # In our build_hsn_graph, we stored: title=f"{row['Description']} (Policy: {row['FinalHSN']})"
                 policy_match = re.search(r'\(Policy: (.*?)\)', desc)
                 policy = policy_match.group(1) if policy_match else "Not specified"
                 clean_desc = desc.split(" (Policy:")[0]

                 return f"""
✅ **Direct Code Lookup**

**HSN Code:** {code}
**Description:** {clean_desc}
**Trade Status:** {policy}

**Full Hierarchy:**
{path_str}
"""

        # 1. Check for Broad Category Query (TC3) - REFINED LOGIC
        # Only trigger for very generic queries with classification keywords
        query_lower = user_query.lower()
        is_generic = (
            len(user_query.split()) <= 3 and
            ("classification" in query_lower or "overview" in query_lower) and
            any(word in query_lower for word in ["rubber", "products", "goods"])
        )

        if is_generic:
            # Find the most relevant chapter
            results = self.vector_store.similarity_search(user_query, k=1)
            if results:
                chapter_code = results[0].metadata['chapter']
                chapter_desc = self.graph.nodes[chapter_code]['title']
                return f"""
📚 **Chapter Overview**

**Chapter {chapter_code}:** {chapter_desc}

This chapter covers various products. To provide an accurate 8-digit HSN code, please provide more specific details about the product (e.g., material, usage, or state).
"""

        # 2. Semantic Search with Disambiguation (TC1, TC2, TC4)
        result = retrieve_with_disambiguation(user_query, self.vector_store, self.gemini_client)

        if result['type'] == 'direct_answer':
            code = result['hsn_code']
            path_str = self.get_hierarchy_text(code)

            # Get policy from graph
            node_data = self.graph.nodes[code]
            desc_with_policy = node_data.get('title', '')
            policy_match = re.search(r'\(Policy: (.*?)\)', desc_with_policy)
            policy = policy_match.group(1) if policy_match else "Not specified"

            return f"""
✅ **Best Match Found**

**HSN Code:** {code}
**Description:** {result['document'].metadata['description']}
**Trade Status:** {policy}

**Full Hierarchy:**
{path_str}
"""
        elif result['type'] == 'disambiguation':
            opts = "\n".join([f"- **HSN {o['code']}**: {o['desc']}" for o in result['options']])
            return f"""
🤔 **Ambiguous Request**

I found multiple matches. Please clarify:
**{result['question']}**

**Options:**
{opts}
"""
        else:
            return "❌ No relevant HSN codes found."

    def get_hierarchy_text(self, hsn_code):
        """
        Traverse the graph and return formatted hierarchy text.
        """
        if hsn_code not in self.graph:
            return "Hierarchy not found in graph."

        # We know the structure (2->4->6->8)
        parts = [hsn_code[:2], hsn_code[:4], hsn_code[:6], hsn_code]

        output = []
        for p in parts:
            if p in self.graph.nodes:
                title = self.graph.nodes[p]['title']
                # Clean policy from title for hierarchy display
                clean_title = title.split(" (Policy:")[0]
                level_name = {0:"Chapter", 1:"Heading", 2:"Subheading", 3:"Code"}.get(self.graph.nodes[p]['level'], "Node")
                output.append(f"{level_name} ({p}): {clean_title}")

        return "\n".join([f"{'  '*i}→ {line}" for i, line in enumerate(output)])
```

# 4.3 Advanced Gemini Query with Google Search (Enhancement)

```python
# For complex queries that may benefit from external search

def advanced_gemini_query(query_text, use_search=False, thinking_level="LOW"):
    """
    Advanced query using Gemini with optional Google Search and thinking.

    Args:
        query_text: The query to send to Gemini
        use_search: Whether to enable Google Search tool
        thinking_level: "LOW", "MEDIUM", or "HIGH" for reasoning depth (Default: LOW)

    Returns:
        Response text from Gemini
    """
    contents = [
        types.Content(
            role="user",
            parts=[types.Part.from_text(text=query_text)],
        ),
    ]

    # Configure tools (optional Google Search)
    tools = []
    if use_search:
        tools = [types.Tool(googleSearch=types.GoogleSearch())]

    # Configure generation using the helper
    generate_content_config = get_gen_config(GEMINI_MODEL, thinking_level=thinking_level)
    if use_search:
        generate_content_config.tools = tools

    response_text = ""
    try:
        for chunk in gemini_client.models.generate_content_stream(
            model=GEMINI_MODEL,
            contents=contents,
            config=generate_content_config,
        ):
            if hasattr(chunk, 'text') and chunk.text:
                response_text += chunk.text

        return response_text
    except Exception as e:
        logger.error(f"Advanced Gemini query failed: {str(e)}")
        return f"Error: {str(e)}"

# Example usage (uncomment to test):
# result = advanced_gemini_query(
#     "What are the export policies for natural rubber latex in India?",
#     use_search=True,
#     thinking_level="HIGH"
# )
# print(result)

logger.info("✓ Advanced Gemini query function ready!")
```

```python
# Diagnostic function to inspect retrieval results
def diagnose_query(query, vector_store, k=10):
    """
    Shows what the vector store retrieves for a given query.
    Useful for debugging retrieval issues.
    """
    print(f"🔍 Diagnosing query: '{query}'\n")
    results = vector_store.similarity_search_with_score(query, k=k)

    print(f"Top {min(k, len(results))} results:\n")
    for i, (doc, score) in enumerate(results[:k], 1):
        print(f"{i}. HSN {doc.metadata['hsn_code']} (Score: {score:.4f})")
        print(f"   Subheading: {doc.metadata['subheading']}")
        print(f"   Description: {doc.metadata['description']}")
        print(f"   Context: {doc.metadata.get('full_context', 'N/A')[:80]}...")
        print()

print("✓ Diagnostic function ready!")
print("\n💡 Usage: diagnose_query('natural rubber latex', vector_store, k=5)")
```
